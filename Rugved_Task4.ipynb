```python
x=df.drop('CO2 Emissions(g/km)',axis=1)
y=df['CO2 Emissions(g/km)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
num_cols=['Engine Size(L)','Cylinders','Fuel Consumption City (L/100 km)','Fuel Consumption Hwy (L/100 km)',
            'Fuel Consumption Comb (L/100 km)','Fuel Consumption Comb (mpg)']

cat_cols=['Make','Model','Vehicle Class','Transmission','Fuel Type']
```

```python
num_pipeline= Pipeline([
    ('imputer',SimpleImputer(strategy='mean')),
    ('scaler',StandardScaler())])


cat_pipeline= Pipeline([
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('encoder',OneHotEncoder(handle_unknown='ignore',sparse_output=False))
])

preprocessor=ColumnTransformer([
    ('num',num_pipeline,num_cols),
    ('cat',cat_pipeline,cat_cols)
])
```

```python
x_train_processed=preprocessor.fit_transform(x_train)
x_test_processed=preprocessor.transform(x_test)
```

```python
n, num_features = x_train_processed.shape
y_train_np = y_train.to_numpy().reshape(-1, 1) if isinstance(y_train, pd.Series) else y_train.reshape(-1, 1)

# weights and bias
weights = np.zeros((num_features, 1))
bias = 0.0

# hyperparameters
learning_rate = 0.1
n_iterations = 500

cost_history = []

# Gradient Descent Loop
for i in range(n_iterations):
    # Calculate predictions
    y_pred = np.dot(x_train_processed, weights) + bias

    #  error finding
    error = y_pred - y_train_np

    # Calculate cost (Mean Squared Error)
    cost = (1/n) * np.sum(error**2)
    cost_history.append(cost)

    # Calculate gradients
    d_weights = (2/n) * np.dot(x_train_processed.T, error)
    d_bias = (2/n) * np.sum(error)

    # Update weights and bias
    weights = weights - learning_rate * d_weights
    bias = bias - learning_rate * d_bias

    if (i+1) % 100 == 0:
        print(f"Iteration {i+1}, Cost: {cost:.4f}")

print(f"\nFinal Weights:\n{weights}")
print(f"Final Bias: {bias}")
print(f"Final MSE (Cost): {cost_history[-1]:.4f}")
```

```python
y_pred_gd = np.dot(x_test_processed, weights) + bias
```

```python
y_pred_gd
```

```python
import numpy as np

y_test_np = y_test.to_numpy().reshape(-1, 1) if isinstance(y_test, (pd.Series, pd.DataFrame)) else y_test.reshape(-1, 1)

# mse= (1/n) * sum((y_test_np - y_pred)^2)
errors = y_test_np - y_pred_gd
squared_errors = errors**2
mse_manual = np.mean(squared_errors)
print(f"MSE: {mse_manual:.4f}")


rmse_manual = np.sqrt(mse_manual)
print(f"RMSE: {rmse_manual:.4f}")
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(y_test_np, y_pred_gd, alpha=0.4)
plt.plot([y_test_np.min(), y_test_np.max()], [y_test_np.min(), y_test_np.max()], 'r--')
plt.xlabel('Actual CO2 Emissions (g/km)')
plt.ylabel('Predicted CO2 Emissions (g/km)')
plt.grid(True)
plt.show()
```
